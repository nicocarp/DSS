```python
class Customer(Base):
    __tablename__ = 'Customer'
    
    code = Column(Integer, primary_key=True)
    name = Column(String)
    address = Column(String)
    phone = Column(String)
    bday = Column(Date)
    gender = Column(String)
    
    #Para formateo de objeto
    def __repr___(self):
        return "<Customer(name='%s', address='%s', phone='%s', gender=%s')>"%(self.name, self.address, self.phone, self.gender)
```

```python
class Timestamp(Base):
    __tablename__ = 'Timestamp'
    
    code = Column(Integer, primary_key=True)
    timestamp = Column(BigInteger)
    date = Column(Date)
    year = Column(Integer)
```

```python
class ClaseVino(Base):
    __tablename__ = 'ClaseVino'
    
    code = Column(Integer, primary_key=True)
    name = Column(String)
    region = Column(String)
```

```python
class Wine(Base):
    __tablename__ = 'Wine'
    
    code = Column(Integer, primary_key=True)
    name = Column(String)
    tipo = Column(String)
    vintage = Column(Integer)
    bottlePrice = Column(Float)
    casePrice = Column(Float)
    claseVino = Column(Integer)
```

```python
class Order(Base):
    __tablename__ = 'Order'
    
    code = Column(Integer, primary_key=True)
    customer = Column(Integer)
    wine = Column(Integer)
    timeStamp = Column(Integer)
    nrBottles = Column(Integer)
    nrCases = Column(Integer)    
```

```python
Base.metadata.create_all(engine)
```

```python
def get_tipo_vinos_mas_vendidos_por_anio(anio):
    assert type(anio) == int, "Error"
    anio = str(anio)
    string_sql = "select w.type as tipo_vino, sum(o.\"nrBottles\") as cant_x_tipo_vino, (select sum(oo.\"nrBottles\") from \"Order\" as oo join \"Timestamp\" as tt on oo.time = tt.code and tt.year = "+anio+") as total_anual, (((sum(o.\"nrBottles\"))*100)/(select sum(oo.\"nrBottles\") from \"Order\" as oo join \"Timestamp\" as tt on oo.time = tt.code and tt.year = "+anio+")) as porcentaje, '"+anio+"' as anio from \"Wine\" as w join \"Order\" as o on (o.wine = w.code) join \"Timestamp\" as t on (o.time = t.code and t.year = "+anio+") group by w.type order by cant_x_tipo_vino desc;"
    resultado  = pd.read_sql(string_sql, engine)
    print "Muestre porcentajes de tipos de vinos m as vendidos en el anio "+anio 
    return resultado

get_tipo_vinos_mas_vendidos_por_anio(2004)
    
```

```python
# Cu al es la temporada con mayor cantidad de ventas de X vino?

def get_temporada_mayor_ventas_por_vino(code_vino):
    string_sql = 'select * from \"Order\" where wine = %d;'%code_vino    
    df_ventas = pd.read_sql(string_sql, engine)
    df_time = pd.read_sql("select * from \"Timestamp\"", engine)    
    
    ventas_por_anio = {}
    for time, n_bottles in df_ventas[['time', 'nrBottles']].values:
        result = df_time.loc[df_time['code'] == time]
        year =  int(result['year'])
        try:
            ventas_por_anio[year] += n_bottles
        except:
            ventas_por_anio[year] = n_bottles
        
    maximo = max(ventas_por_anio.values())
    resultado = filter(lambda (k, v): True if v == maximo else False, ventas_por_anio.items())
    anios = map(lambda (k,v):  k, resultado)
    return maximo, anios

max_vendido, anios = get_temporada_mayor_ventas_por_vino(10)
print "Cantidad maxima %d En los anios %s "%(max_vendido, anios)

    
```
